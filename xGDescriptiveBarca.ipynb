## Expected Goals (xG) Brief Definition

Expected Goals is a statistical measurement that was created with the aim of getting a more accurate reflection of what really happened during a game in relation to the quality of opportunities that each team was able to create. Unlike traditional stats (total goals scored, total shots, shots on target, etc.), xG focuses on the quality of the goals/opportunities rather than the quantity, and assigns a probability (based on historical data) for each shot that is performed.

Each goal opportunity will contribute to a point system. For example, a penalty kick supposes a 76% of probability of being scored, that is 0.76 points that will be added to the other xG values of other shots made by the same team.

#### Example: Peru vs Denmark WC 2018

Peru 0-1 Denmark. The game is well remembered by Peruvians as one of the most dissapointing matches in recent years. The excitement of qualifying for the World Cup after 36 years had been hit hard by the defeat at the hands of Yusuf Poulsen with a goal in the 58th minute. However, many Peruvians (myself included) were left with the feeling of having competed well and being far superior to the rival. Let's interpret the xG data from that match.


```python
# importing pandas as reading the dataset into the pd environment
import pandas as pd 

df1 = pd.read_csv("Peru_vs_Den_xGShots.csv")

# leveraging groupby function for better visualization
df1.groupby(by=["team", "player"])[["shot_statsbomb_xg"]].sum()
```

(Data extracted using statsbombpy package). The table depicts the probability of scoring for each player on each team. As mentioned before, we can see that Christian Cueva had a total xG value of 0.76 (76%) because he missed a penalty right before the end of the first half. 

Each of these xG values will be added to the Overall xG value for each team: 

```python
# groupby with sum function to get the overall xG sum() per team

df1.groupby(by=["team"])[["shot_statsbomb_xg"]].sum()
```

In this case, the xG data confirms that Peru did have better opportunities of scoring based on probabilities, and that the final score was not a real reflection of what happened during the game. 

## Xavi's appointment as manager of FC Barcelona

On November 6th 2021, Xavi's was announced as the new head coach of FC Barcelona's first team given the club's poor performances under the former manager and club legend Ronald Koeman. 

Xavi's arrival was presented as a commitment by the sports management and president Joan Laporta to return to the foundations that made Barca great at the time: the Cruyff school. Xavi Hernandez is a loyal believer in offensive and positional football, philosophically aligning himself with important figures from Barca and football history such as Cruyff, Guardiola and Luis Enrique. It is precisely this ideological relationship, added to the fact that Xavi is a living legend of the club and was part of Guardiola's golden age (2008-2012), sufficient reasons for the illusion of the fans to prevail over perhaps a lack of experience at the highest level as a coach. However, from an idiosyncratic point of view, there were very few who had the requirements to fulfill this role as much as Xavi.

From a tactical point of view, Xavi has strengthened the squad in aspects such as control and intensity of the ball in circulation, with "Cruyffist" concepts such as the constant search for the third man, and the intensity and importance of a good defensive build up of the ball. Also, one of the most important aspects, the intensity in high pressure, a fundamental pillar of the great Barca of the past, was recovered.

This change is evident in the improvement in the performance of Piqué, Alba, Eric García, Busquets, and many others who will be the subject of study in future analyses. For the moment, let's see the descriptive statistics that Xavi's Barca has left us in these 145 days.

```python
# import packages and dataset
import pandas as pd
import numpy as np
import seaborn as sns

xg_df = pd.read_excel("xGBarcelona.xlsx")

# make two separate dataFrame for the descriptive statistics section, one for "before Xavi's arrival" and another for "after Xavi's arrival"
non_xavi_df = xg_df.loc[xg_df["Coach"] != "Xavi"]
xavi_df = xg_df.loc[xg_df["Coach"] == "Xavi"]
```

## 1. Average xG per Game

```python
# Overall mean in both Home and Away games
fcb_xG_before = non_xavi_df["xG FCB"].mean()
fcb_xG_after = xavi_df["xG FCB"].mean()

graph_1 = sns.catplot(data=xg_df, x="Coach", y="xG FCB", kind="bar", palette="Blues")
# graph_1.set_xticklabels(rotation=90)

print("Average FCB xG per Game (before Xavi): " + str(non_xavi_df["xG FCB"].mean()))
print("Average FCB xG per Game (with Xavi): " + str(xavi_df["xG FCB"].mean()))

increase = (xavi_df["xG FCB"].mean() - non_xavi_df["xG FCB"].mean()) / non_xavi_df["xG FCB"].mean() * 100
print("Percentage increase per game of: " + str(round(increase, 2)) + "%, quantifying as " + str(xavi_df["xG FCB"].mean() - non_xavi_df["xG FCB"].mean()) + 
      " total xG value increase per game on average.")
```

### Comments: 

Although it may not seem like much, an average increase of 0.51 xG per match is a significant improvement. Xavi's Barcelona now enjoys producing an extra 1/2 goal per game based on probability, with respect to Barcelona xG values before his arrival. In point 5, we will be able to compare the average xG of other top teams in Europe and understand the magnitude of this improvement.

## 2. Average xG per Game (Camp Nou vs Away)

```python
# xG statistics as Hosts and Away games

graph_2 = sns.catplot(data=xg_df, x="Coach", y="xG FCB", kind="bar", hue="H/A", palette="Accent")
# graph_2.set_xticklabels(rotation=90)
```

```python
xg_df.groupby(by=["Coach", "H/A"])[["xG FCB"]].mean()   # get xG values
# assign xG to the following variables (still need to find a way to do it programatically)
koeman_away_xG = 1.287500
koeman_home_xG = 2.245000
xavi_away_xG = 2.285000
xavi_home_xG = 2.291667

print("Xavi's FCB is able to produce an increase of " + str(xavi_home_xG - koeman_home_xG) + " in xG values at HOME per game (" + 
       "{:.2f}".format((xavi_home_xG - koeman_home_xG)/ koeman_home_xG*100) + "% increase)")
print("Xavi's FCB is able to produce an increase of " + str(xavi_away_xG - koeman_away_xG) + " in xG values AWAY per game (" + 
       "{:.2f}".format((xavi_away_xG - koeman_away_xG)/ koeman_away_xG*100) + "% increase)")
```

### Comments: 

What is striking about these statistics is not only that Xavi has managed to maintain similar xG averages both at home and away (which shows off a special consistency), but also that he has improved Koeman's away xG by 77.48%. This fact condemns the performance of Koeman's Barca, who leveraged the advantages of playing at the Camp Nou, but failed to create great scoring opportunities outside of it, unlike Xavi, who maintains consistent numbers. We can clearly see the difference in competitiveness that the team has experienced throughout this season.

## 3. Average xG difference to the opposition

```python
# Average difference among coaches
graph_3 = sns.catplot(data=xg_df, x="Coach", y="xG Difference", kind="box", palette="BrBG")
graph_3.fig.set_size_inches(10,5)
graph_3.fig.subplots_adjust(top=0.81, right=0.80)

xavi_df_diff = xavi_df["xG Difference"].mean()
non_xavi_df_diff = non_xavi_df["xG Difference"].mean()

print("Xavi's FCB enjoys a xG difference value with respect to the opposition's of: " + str(xavi_df_diff) + ", an increase of " +
      "{:.2f}".format((xavi_df_diff - non_xavi_df_diff)/ non_xavi_df_diff*100) + "% to the previous average difference of (" + str(non_xavi_df_diff) + ")")
```

### Comments: 

These numbers show the ability of Xavi's Barca to not only enjoy a higher quality of opportunities than the rival, but also to have improved the difference to xG with respect to the rival by 86.20%, on average. With xG difference we understand as the average difference in xG score of Barca with respect to its rivals.

##### Example: Camp Nou - March 12, 2022
Barca (3.9 xG) vs (0.11 xG) Osasuna

The difference in xG in this case would be 3.79 in favor of Barca (3.9 - 0.11), which symbolizes the superiority of one team over the other.
